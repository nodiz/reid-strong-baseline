```python
import glob
import random
import os
import sys
import math
import numpy as np
from PIL import Image
import torch
import torchvision.transforms as transforms
basepath = "../../"
car_imgs = sorted(glob.glob(os.path.join(basepath, "cars_/*.*")))
len_imgs = len(car_imgs)

print(car_imgs)
print(len_imgs)
```

```python
# img preprocessing

import matplotlib.pyplot as plt
%matplotlib inline

car = car_imgs[1]
img_car = np.asarray(Image.open(car).convert('RGB'))/255

# Only keeping down half and removing borders
h,w = img_car.shape[:2]
img_car = img_car[h//2:h//10*9, w//20:w//20*19, :]
img_h,img_w = img_car.shape[:2]
plt.imshow(img_car)
```

```python
# Crop a random part receiving h, w = height, wdith
h = 100
w = 200

y1 = random.randint(0, img_h - h)
x1 = random.randint(0, img_w - w)
print(img_car.shape)
print(x1,y1,w,h)
img_car = img_car[y1:y1+h, x1:x1+w,:]
print(img_car.shape)
plt.imshow(img_car)
```

```python
# Transpose
print(img_car[10,4,:])
img_car = np.transpose(img_car, (2, 0, 1))
print(img_car[:,10,4])
print(img_car.shape)
```
